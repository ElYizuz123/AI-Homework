```python
from transformers import pipeline

clasificador = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

def clasificar_comentario(texto):
    resultado = clasificador(texto)[0]
    label = resultado["label"]

    # Etiquetas del modelo: 1 estrella (malo) → 5 estrellas (bueno)
    if "1 star" in label or "2 stars" in label:
        return "Comentario MALO"
    elif "4 stars" in label or "5 stars" in label:
        return "Comentario BUENO"
    else:
        return "Neutral"

print(clasificar_comentario("Odio este servicio"))
print(clasificar_comentario("Me encantó todo"))
```
