```python
import pandas as pd
from transformers import pipeline

# Cargamos el clasificador
clasificador = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

def clasificar_comentario(texto):
    resultado = clasificador(texto)[0]
    label = resultado["label"]

    # Etiquetas del modelo: 1 estrella (malo) â†’ 5 estrellas (bueno)
    if "1 star" in label or "2 stars" in label:
        return "MALO"
    elif "4 stars" in label or "5 stars" in label:
        return "BUENO"
    else:
        return "NEUTRAL"


df = pd.read_csv("opiniones_frankenstain.csv")  

# Clasificar cada comentario
df["clasificacion"] = df["Comentario_Reaccion"].apply(clasificar_comentario)

# Calcular porcentajes
total = len(df)
porcentaje_buenos = (df["clasificacion"].value_counts().get("BUENO", 0) / total) * 100
porcentaje_malos = (df["clasificacion"].value_counts().get("MALO", 0) / total) * 100
porcentaje_neutros = (df["clasificacion"].value_counts().get("NEUTRAL", 0) / total) * 100

# Mostrar resultados
print("RESULTADOS GENERALES:")
print(f"Comentarios BUENOS: {porcentaje_buenos:.2f}%")
print(f"Comentarios MALOS: {porcentaje_malos:.2f}%")
print(f"Comentarios NEUTRALES: {porcentaje_neutros:.2f}%")

# Guardar el CSV con la nueva columna
df.to_csv("comentarios_clasificados_frankenstain.csv", index=False)
print("\nArchivo generado: comentarios_clasificados_frankenstain.csv")
```

```python
import pandas as pd
from transformers import pipeline
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans

print(" Cargando modelo de clasificaciÃ³n...")
clasificador = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

print(" Cargando modelo de embeddings...")
embedder = SentenceTransformer('all-MiniLM-L6-v2')

print(" Cargando modelo de resumen...")
resumidor = pipeline("summarization", model="facebook/bart-large-cnn")


def clasificar_comentario(texto):
    try:
        resultado = clasificador(texto)[0]
        label = resultado["label"]

        if "1 star" in label or "2 stars" in label:
            return "MALO"
        elif "4 stars" in label or "5 stars" in label:
            return "BUENO"
        else:
            return "NEUTRO"
    except:
        return "NEUTRO"


RUTA_CSV = "opiniones_gen_z.csv"  # Cambiar por tu archivo
df = pd.read_csv(RUTA_CSV)

# Asegurate de tener una columna "comentario"
if "Comentario_Reaccion" not in df.columns:
    raise Exception("Tu CSV debe contener una columna llamada 'Comentario_Reaccion'.")

print(" Clasificando comentarios...")

df["clasificacion"] = df["Comentario_Reaccion"].astype(str).apply(clasificar_comentario)


total = len(df)
buenos = len(df[df["clasificacion"] == "BUENO"])
malos = len(df[df["clasificacion"] == "MALO"])

porc_buenos = (buenos / total) * 100
porc_malos = (malos / total) * 100

print("\n RESULTADOS DE CLASIFICACIÃ“N:")
print(f" Comentarios buenos: {porc_buenos:.2f}% ({buenos}/{total})")
print(f" Comentarios malos: {porc_malos:.2f}% ({malos}/{total})")



print("\n Generando embeddings para agrupar ideas...")
comentarios = df["Comentario_Reaccion"].astype(str).tolist()
embeddings = embedder.encode(comentarios, show_progress_bar=True)


n_clusters = 5  
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df["cluster"] = kmeans.fit_predict(embeddings)

print(" Clusters creados:", n_clusters)


def resumir_texto(textos):
    contenido = " ".join(textos)
    if len(contenido.split()) < 30:
        return contenido  
    resumen = resumidor(contenido, max_length=80, min_length=30, do_sample=False)[0]["summary_text"]
    return resumen


print("\nðŸ“ RESÃšMENES POR GRUPO:")
for i in range(n_clusters):
    grupo = df[df["cluster"] == i]["Comentario_Reaccion"].tolist()
    print(f"\nðŸ”· Cluster {i} ({len(grupo)} comentarios):")

    try:
        resumen = resumir_texto(grupo)
        print("Resumen:", resumen)
    except:
        print("No se pudo resumir este grupo.")


df.to_csv("comentarios_clasificados_y_agrupados_gen_z.csv", index=False)
print("\n Archivo generado: comentarios_clasificados_y_agrupados_gen_z.csv")
```
