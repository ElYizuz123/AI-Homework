```python
import pandas as pd
from transformers import pipeline

# Cargamos el clasificador
clasificador = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

def clasificar_comentario(texto):
    resultado = clasificador(texto)[0]
    label = resultado["label"]

    # Etiquetas del modelo: 1 estrella (malo) ‚Üí 5 estrellas (bueno)
    if "1 star" in label or "2 stars" in label:
        return "MALO"
    elif "4 stars" in label or "5 stars" in label:
        return "BUENO"
    else:
        return "NEUTRAL"


df = pd.read_csv("opiniones_frankenstain.csv")  

# Clasificar cada comentario
df["clasificacion"] = df["Comentario_Reaccion"].apply(clasificar_comentario)

# Calcular porcentajes
total = len(df)
porcentaje_buenos = (df["clasificacion"].value_counts().get("BUENO", 0) / total) * 100
porcentaje_malos = (df["clasificacion"].value_counts().get("MALO", 0) / total) * 100
porcentaje_neutros = (df["clasificacion"].value_counts().get("NEUTRAL", 0) / total) * 100

# Mostrar resultados
print("RESULTADOS GENERALES:")
print(f"Comentarios BUENOS: {porcentaje_buenos:.2f}%")
print(f"Comentarios MALOS: {porcentaje_malos:.2f}%")
print(f"Comentarios NEUTRALES: {porcentaje_neutros:.2f}%")

# Guardar el CSV con la nueva columna
df.to_csv("comentarios_clasificados_frankenstain.csv", index=False)
print("\nArchivo generado: comentarios_clasificados_frankenstain.csv")
```

```python
import pandas as pd
from transformers import pipeline
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans

print("üîÑ Cargando modelo de clasificaci√≥n...")
clasificador = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

def clasificar_comentario(texto):
    resultado = clasificador(texto)[0]
    label = resultado["label"]

    if "1 star" in label or "2 stars" in label:
        return "MALO"
    elif "4 stars" in label or "5 stars" in label:
        return "BUENO"
    else:
        return "NEUTRAL"

# Cargar dataset
df = pd.read_csv("opiniones_frankenstain.csv") 

print("üè∑Ô∏è Clasificando comentarios...")
df["clasificacion"] = df["Comentario_Reaccion"].apply(clasificar_comentario)

# Estad√≠sticas
total = len(df)
porc_buenos = df["clasificacion"].value_counts().get("BUENO", 0) / total * 100
porc_malos  = df["clasificacion"].value_counts().get("MALO", 0) / total * 100
porc_neutros = df["clasificacion"].value_counts().get("NEUTRAL", 0) / total * 100

print("\nüìä RESULTADOS:")
print(f"BUENOS: {porc_buenos:.2f}%")
print(f"MALOS: {porc_malos:.2f}%")
print(f"NEUTRALES: {porc_neutros:.2f}%")

df.to_csv("comentarios_clasificados.csv", index=False)


#Agrupamiento de ideas.

print("\nüîÑ Cargando modelo de embeddings...")
modelo_embed = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

print("üß† Generando embeddings...")
embeddings = modelo_embed.encode(df["Comentario_Reaccion"], show_progress_bar=True)

num_clusters = 5  # puedes modificarlo

print("üìå Agrupando ideas...")
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df["cluster"] = kmeans.fit_predict(embeddings)

df.to_csv("comentarios_agrupados.csv", index=False)


# Resumen general

print("\nüîÑ Cargando modelo de resumen...")
resumidor = pipeline("summarization", model="facebook/bart-large-cnn")

texto_completo = " ".join(df["comentario"].tolist())[:3000]

print("\nüìù Generando resumen general...")
resumen_general = resumidor(texto_completo, max_length=150, min_length=60, do_sample=False)

print("\nüìÑ RESUMEN GENERAL:")
print(resumen_general[0]["summary_text"])


# Resumen por clusters

print("\nüì¶ RES√öMENES POR GRUPO:")

for c in range(num_clusters):
    subset = df[df["cluster"] == c]["Comentario_Reaccion"].tolist()
    
    if len(subset) < 3:
        continue
    
    texto_cluster = " ".join(subset)[:3000]
    resumen_cluster = resumidor(texto_cluster, max_length=120, min_length=40, do_sample=False)

    print(f"\nüìå CLUSTER {c}:")
    print(resumen_cluster[0]["summary_text"])

    print("\nEjemplos representativos:")
    print(df[df["cluster"] == c]["Comentario_Reaccion"].head(5).to_string(index=False))


print("\n‚úÖ PROCESO COMPLETO FINALIZADO.")
print("üìÅ Archivos generados:")
print(" - comentarios_clasificados.csv")
print(" - comentarios_agrupados.csv")
```
