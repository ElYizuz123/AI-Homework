# Dependencias

```python
from sklearn import tree
import graphviz
import pandas as pd
import joblib
```

# Cargar el conjunto de datos Iris

```python
data = pd.read_csv("emociones.csv")
X = data.drop(columns=["clase"])
y = data["clase"]
```

# Crear el clasificador del Árbol de Decisión

```python
clf = tree.DecisionTreeClassifier()
```

# Entrenar el modelo con los datos

```python
clf = clf.fit(X, y)
```

# Exportar el árbol de decisión en formato DOT para su visualización

```python
dot_data = tree.export_graphviz(
    clf,
    out_file=None,
    feature_names=X.columns,         
    class_names=["Neutro", "Feliz", "Triste", "Enojado", "Sorprendido"], 
    filled=True,
    rounded=True,
    special_characters=True
)
```

# Crear el gráfico con graphviz

```python
graph = graphviz.Source(dot_data)
```

# Guardar el gráfico como un archivo PDF (opcional)

```python
graph.render("modelo_emociones")
```

```python
# Mostrar el gráfico directamente
```

```python
graph.view()
```

```python
joblib.dump(clf, "modelo_emociones.pkl")
```
